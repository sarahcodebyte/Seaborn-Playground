# **Importing dependencies**

```python
!pip install seaborn
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Data collection and analysis**

```python
dataset = pd.read_csv('/content/data.csv')
```

# **Statistical Relationship_ScatterPlot**

```python
sns.relplot(x='study_hours', y='student_marks', hue='age', data=dataset)
```

# **Statistical Relationship_LinePlot**

```python
sns.relplot(x='study_hours', y='student_marks', hue='age', data=dataset, kind = 'line', style = 'internet')
```

# **Statistical Relationship_Multiple**

```python
sns.relplot(x='study_hours', y='student_marks', hue='age', data=dataset, col='internet')
```

# **Data Distribution_Histogram**

```python
sns.displot(dataset, x='age', binwidth=0.5, hue='internet')
```

# **Data Distribution_Multiple**

```python
sns.displot(dataset, x='student_marks', col='internet')
```

# **Data Distribution_Bivariate**

```python
sns.displot(dataset, x='age', y='student_marks', hue='internet')
```

# **Data Distribution_Kernel Density**

```python
sns.displot(dataset, x='age', y = 'student_marks', kind='kde', hue='internet')
```

# **Data Distribution_PairPlot**

```python
sns.pairplot(dataset)
```

# **Categorical Data_Scatterplot**

```python
sns.catplot(x='age', y='student_marks', data=dataset, hue='internet')
```

# **Categorical Data_Box Plot**

```python
sns.catplot(x='age', y='student_marks', kind='box', data=dataset, hue='internet')
```

# **Regression Output**

```python
sns.regplot(x = 'student_marks', y = 'study_hours', data=dataset)
sns.lmplot(x='student_marks', y='study_hours', data=dataset, hue='internet', markers=['o', 'x'], palette='Set1')
```

# **Using SNS theme in other graph**

```python
def sinplot(flip=1):
  x = np.linspace(0, 14, 100)
  for i in range(1, 7):
    plt.plot(x, np.sin(x + i * .5)*(7 - i) * flip)

sns.set_context('talk') #talk/paper/poster
sinplot()    
```
